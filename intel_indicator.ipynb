```
#Import Libs
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```
#load intel data
from google.colab import files
upload = files.upload()
```

```
# store the data
INTC = pd.read_csv('INTC.csv')
#Show data
INTC
```

```
#Visual the data
plt.figure(figsize=(12.5, 4.5))
plt.plot(INTC['Adj Close'], label = 'Intel')
plt.title('Intel Adj. Close Price')
plt.xlabel('July 2019 - November 2020')
plt.ylabel('Adj. Close Price USD ($)')
plt.legend(loc='upper left')
plt.show()
```

```
#Create Simple moving avarage with 30 days.
SMA30 = pd.DataFrame()
SMA30['Adj Close'] = INTC['Adj Close'].rolling(window= 30).mean()
SMA30
```

```
#Create SMA with a 100 days 
SMA100 = pd.DataFrame()
SMA100['Adj Close'] = INTC['Adj Close'].rolling(window=100).mean()
SMA100
```

```
#visual the data
plt.figure(figsize=(12.5, 4.5))
plt.plot(INTC['Adj Close'], label = 'Intel')
plt.plot(SMA30['Adj Close'], label = 'SMA30')
plt.plot(SMA100['Adj Close'], label = 'SMA100')
plt.title('Intel Adj. Close Price')
plt.xlabel('July 2019 - November 2020')
plt.ylabel('Adj. Close Price USD ($)')
plt.legend(loc='upper left')
plt.show()
```
